```python
# Définition des fonctions que je vais utiliser
# Fonction pour définir le type de chaque DataFrame (manuellement a défaut de comprendre comment le faire avec du code)
def define_type(name):
    if (name == "data_statsCountrySeries"):
        print(f"Référentiel des indicateurs (Series Code, Topic).")
    elif (name == "data_statsCountry"):
        print(f"Référentiel des pays (Country Code).")
    elif (name == "data_statsData"):
        print(f"Il s'agit du dataframe des données principales (Country Code + Indicator Code + années).")
    elif (name == "data_statsFootnote"):
        print(f"Il s'agit du dataframe des notes/annotations (Series Code).")
    elif (name == "data_statsSeries"):
        print(f"Liens pays-indicateurs (Country Code + Series Code).")

# Fonction pour vérifier les doublons dans chaque DataFrame
def check_doublons(df):
    # On vérifie les doublons
    duplicates = df[df.duplicated()] # On crée une copie du DataFrame pour éviter de modifier l'original
    if not duplicates.empty:
        print(f"Il y a {duplicates.shape[0]} doublons.") # Affiche le nombre de doublons
        print(duplicates)  # Affiche les doublons trouvés
        df.drop_duplicates(inplace=True) # On supprime les doublons
        # On reverifie qu'il n'y a plus de doublons
        if df.duplicated().any():
            print(f"Il reste des doublons.")
        else:
            print(f"Tous les doublons ont été supprimés.")
    else:
        print(f"Il n'y a pas de doublons.")


# Fonction qui calcule la proportion de valeurs manquantes par colonne et on supprime les colonnes vide
def check_columns_isnull(df):
    missing_counts = df.isnull().sum()
    total_counts = df.shape[0] # Calculer et afficher la proportion de valeurs manquantes par colonne
    missing_percentage = (missing_counts / total_counts * 100).sort_values(ascending=False)
    display(missing_percentage.to_frame(name="Pourcentage de valeurs manquantes"))
    if missing_percentage.any() > 0:
        print("Certaine(s) colonnes sont vide(s), elle(s) va(vont) être supprimée(s).")
        df.dropna(axis=1, how='all', inplace=True) # Suppression des colonnes avec 100% de valeurs manquantes
    else:
        print("Aucune colonne vide.")

# Fonction qui calculez les statistiques descriptives basiques
def describe_data(df):
    # Afficher les statistiques descriptives avec arrondi
    print("Statistiques descriptives :")
    display(df.describe().round(2))  # Arrondit les valeurs à 2 décimales
```

```python
# On associe chaque DataFrame à son nom pour plus de lisibilité
all_data_frames = {
    "data_statsCountrySeries": data_statsCountrySeries,
    "data_statsCountry": data_statsCountry,
    "data_statsData": data_statsData,
    "data_statsFootnote": data_statsFootnote,
    "data_statsSeries": data_statsSeries
}

# Boucle qui parcourt chaque DataFrame et donne les informations demandées
for name, df in all_data_frames.items():
    print(f"###################################################### {name} ######################################################") # On affiche le nom du DataFrame
    define_type(name) # On définit le type du DataFrame
    print(f"Il y a : {df.shape[0]} lignes et {df.shape[1]} colonnes AVANT le nettoyage")  # Affiche le nombre de lignes et de colonnes dans chaque DataFrame avant le nettoyage
    check_doublons(df) # On vérifie les doublons dans le DataFrame et on les supprime
    check_columns_isnull(df)  # On vérifie les colonnes avec des valeurs manquantes et on supprime les colonnes vides
    print(f"Il y a : {df.shape[0]} lignes et {df.shape[1]} colonnes APRES le nettoyage")  # Affiche le nombre de lignes et de colonnes dans chaque DataFrame après le nettoyage
    describe_data(df)  # Affiche les statistiques descriptives du DataFrame
    print("\n")  # Ajoute une ligne vide pour séparer les DataFrames
```

3. Identification et suppression des indicateurs redondants

```python
def identify_redundant_indicators(correlation_matrix):
    # On considère une corrélation > 0.9 ou < -0.9 comme redondante
    threshold = 0.9
    redundant_pairs = []

    # Parcourir la matrice de corrélation pour trouver les paires redondantes
    for i in correlation_matrix.columns:
        for j in correlation_matrix.columns:
            if i != j and (correlation_matrix.loc[i, j] > threshold or correlation_matrix.loc[i, j] < -threshold):
                redundant_pairs.append((i, j, correlation_matrix.loc[i, j]))

    # Afficher les paires redondantes
    print("Indicateurs fortement corrélés (redondants) :")
    for pair in redundant_pairs:
        print(f"{pair[0]} et {pair[1]} : corrélation = {pair[2]:.2f}")
    
    return redundant_pairs
```

* Suppression des indicateurs redondants

```python
# Identification des indicateurs redondants
redundant_pairs_pearson = identify_redundant_indicators(correlation_matrix_pearson)

# On garde uniquement un indicateur par paire redondante
indicators_to_remove = set()
for pair in redundant_pairs_pearson:
    indicators_to_remove.add(pair[1])  # Supposons que nous supprimons le deuxième indicateur de chaque paire

# Filtrer les colonnes pour ne garder que les indicateurs non redondants
data_countries_topics_filtered = data_countries_topics.drop(columns=indicators_to_remove, errors='ignore')

# Afficher les indicateurs restants
print("\nIndicateurs restants après suppression des redondances :")
print(data_countries_topics_filtered.columns.tolist())
```

#### **Analyse des indicateurs et classement des pays :**

```python
# Résumé par indicateur: couverture (nb de pays non-nuls), moyenne et écart-type
summary_rows = []
for col in data_restante_after_pearson.columns:
    s = data_restante_after_pearson[col]
    coverage = s.notna().sum()
    mean = s.mean()
    std = s.std()
    summary_rows.append({"Indicateur": col, "Couverture": coverage, "Moyenne": mean, "Ecart-type": std})

summary_df = pd.DataFrame(summary_rows).sort_values(by=["Couverture","Moyenne"], ascending=[False, False])
print("Résumé des indicateurs restants (triés par couverture puis moyenne):")
display(summary_df)

# Top et Bottom pays par indicateur (selon valeur)
print("\nTop 3 et Bottom 3 par indicateur:")
for col in data_restante_after_pearson.columns:
    s = data_restante_after_pearson[col].dropna()
    if len(s) == 0:
        continue
    top3 = s.sort_values(ascending=False).head(3)
    bottom3 = s.sort_values(ascending=True).head(3)
    print(f"\nIndicateur: {col}")
    print("  Top 3:")
    for country, val in top3.items():
        print(f"    - {country}: {val:.3f}")
    print("  Bottom 3:")
    for country, val in bottom3.items():
        print(f"    - {country}: {val:.3f}")

# Normalisation min-max par indicateur (0-1) et score composite moyen
norm_data = data_restante_after_pearson.copy()
for col in norm_data.columns:
    s = norm_data[col]
    min_v, max_v = s.min(), s.max()
    if pd.isna(min_v) or pd.isna(max_v) or max_v == min_v:
        # Si pas de variance, mettre à 0.5 (ne discrimine pas)
        norm_data[col] = s.apply(lambda x: 0.5 if pd.notna(x) else np.nan)
    else:
        norm_data[col] = (s - min_v) / (max_v - min_v)

# Calcul du score composite (moyenne des indicateurs normalisés non-nuls)
composite_score = norm_data.mean(axis=1, skipna=True).sort_values(ascending=False)
print("\nTop 5 pays par score composite (plus élevé = mieux):")
print(composite_score.head(5))

# Conserver un DataFrame des scores pour usage ultérieur
country_scores = composite_score.to_frame(name="ScoreComposite")
display(country_scores.head(10))
```
