```python
# On importe les bibliothèques nécessaires
import numpy as np
import pandas as pd
```

```python
# On lit le fichier CSV qui est dans le même dossier que ce notebook
prets = pd.read_csv('./prets.csv')
```

```python
# On ajoute une colonne 'endetement' qui calcule le taux d'endettement
prets['endetement'] = (prets['remboursement']*100 / prets['revenu']).round(2)
```

```python
# On modifie le nom de la colonne 'taux' en 'taux_interet'
prets.rename(columns={'taux': 'taux_interet'}, inplace=True)
```

```python
# On ajoute la colonne 'cout_total' qui calcule le coût total du prêt
prets['cout_total'] = (prets['remboursement'] * prets['duree']).round(2)
```

```python
# On ajoute une colonne 'benefice' qui calcule le bénéfice net du prêt
prets['benefice'] = (((prets['cout_total'] * prets['taux_interet']/100)) / 24).round(2)
```

```python
# on affiche les 5 prêts les plus rentables
prets.nlargest(5, 'benefice')
```

```python
# On met dans une variable 'clients_risque' les clients dont le taux d'endettement est supérieur à 35%
clients_risque = prets[prets['endetement'] > 35]
clients_risque.head()  # Affiche les premières lignes du DataFrame des clients à risque
```

```python
# On met dans une variable 'clients_risque_paris' les clients dont le taux d'endettement est supérieur à 35% de la ville de Paris
clients_risque_paris = prets[(prets['endetement'] > 35) & (prets['ville'] == 'PARIS')]
clients_risque_paris.head()  # Affiche les premières lignes du DataFrame des clients
```

```python
# On ajoute une colonne 'risque' qui indique si le client est à risque ou non
prets['risque'] = np.where(prets['endetement'] > 35, 'Oui', 'Non')
prets.head(20)
```

```python
# On affiche combien de automobiles sont financées par des prêts
nombre_automobiles_financees = prets[prets['type'] == 'automobile'].shape[0]
print(f"Nombre d'automobiles financées par des prêts : {nombre_automobiles_financees}")

# On affiche le coupt moyen total des prêts pour les préts de type 'Automobile'
cout_moyen_automobile = prets[prets['type'] == 'automobile']['cout_total'].mean().round(2)
print(f"Coût moyen total des prêts pour les prêts de type 'automobile' :{cout_moyen_automobile} €")
```

```python
# On affiche le bénéfice mensuel total pour l'agence de Toulouse
benefice_mensuel_toulouse = prets[prets['ville'] == 'TOULOUSE']['benefice'].sum().round(2)
print(f"Bénéfice mensuel total pour l'agence de Toulouse : {benefice_mensuel_toulouse} €")
```
