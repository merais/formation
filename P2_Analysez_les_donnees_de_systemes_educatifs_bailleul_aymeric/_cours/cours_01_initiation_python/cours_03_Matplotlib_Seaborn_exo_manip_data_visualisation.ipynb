```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# On lit le fichier CSV qui est dans le même dossier que ce notebook
prets = pd.read_csv('./sources/prets_final.txt')
prets.head()  # Affiche les premières lignes du DataFrame prets
```

```python
# On regroupe les données par type de prêt et on compte le nombre de prêts pour chaque type
grp_type = prets.groupby('type')
grp_type['type'].count()

# Avec Matplotlib
#plt.pie(grp_type['type'].count(), labels=grp_type['type'].count().index, autopct='%1.1f%%')
#plt.title('Répartition des types de prêts') 

# Avec Seaborn
sns.countplot(data=prets, x='type')
plt.show()
```

```python
# On crée un graphique à barres pour visualiser les bénéfices par agence
# Avec Matplotlib
#grp_benefices_per_identifiant = prets.groupby(['identifiant','revenu'])['benefices'].sum()
#grp_benefices_per_identifiant = grp_benefices_per_identifiant.reset_index() 
#grp_benefices_per_identifiant = grp_benefices_per_identifiant.sort_values(by='revenu', ascending=False)
#plt.scatter(grp_benefices_per_identifiant['revenu'], grp_benefices_per_identifiant['benefices'], marker='o')
#plt.xlabel('Revenu') 

# Avec Seaborn
sns.scatterplot(data=prets, x='revenu', y='benefices', hue='ville')
plt.show()
```

```python
# On crée un histogramme pour visualiser la distribution des bénéfices
# Avec Matplotlib
#plt.hist(prets['benefices']) 

# Avec Seaborn
sns.histplot(data=prets, x='benefices', kde=True)
```

```python

# Chargement des données du Chiffre d'Affaires
CA = pd.read_csv('./sources/CA.csv')
CA['date'] = pd.to_datetime(CA['date'])
CA.head(20)
```

```python
# Évolution du CA au cours de l'année 2021
plt.figure(figsize=(10, 7))
plt.plot(CA['date'],CA['immobilier'], label='Immobilier', linewidth=3)
plt.plot(CA['date'],CA['automobile'], label='automobile', linewidth=3)
plt.plot(CA['date'],CA['consommation'], label='consommation', linewidth=3)
plt.ylabel('Chiffre d\'Affaires (€)')
plt.title('Évolution du Chiffre d\'Affaires par secteur')
plt.grid()
plt.xticks(['2021-03-01', '2021-06-01', '2021-09-01', '2021-12-01'])
plt.legend(loc='upper right')
plt.yticks(fontsize=11)
plt.show()
```

```python
# création d'une Series de 4 dates
dates = pd.to_datetime(['2021-01-01', '2021-05-01', '2021-09-01', '2022-01-01'])

# création de 3 listes Y1, Y2 et Y3
Y1 = [120, 560, 1760, 1860]
Y2 = [300, 365, 100, 490]
Y3 = [300, 790, 1130, 1400]

plt.figure()
plt.plot(dates, Y1, label='a')
plt.plot(dates, Y2, label='b')
plt.plot(dates, Y3, label='c')
plt.legend()
plt.show()
```
