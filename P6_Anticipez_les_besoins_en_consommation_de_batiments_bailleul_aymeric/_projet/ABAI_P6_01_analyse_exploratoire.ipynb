### 1- On commence par ne garder que les batiments non résidentiels

Ce dataset contient des informations sur la consommation énergétique de bâtiments en 2016. 

**Focus de l'analyse :** Nous nous concentrons uniquement sur les **bâtiments non résidentiels** (commerciaux, bureaux, industriels, etc.) pour analyser leur consommation énergétique et identifier les opportunités d'optimisation.

```python
 # On affiche tout les Builkding Type disponibles
building_types = building_consumption['PrimaryPropertyType'].unique()
print("Types de bâtiments disponibles :")
for b_type in building_types:
    print(f"- {b_type}")
```

```python
# Filtrer pour ne garder que les bâtiments non résidentiels
print(f"Nombre total de bâtiments avant filtrage : {len(building_consumption)}")
initial_count = len(building_consumption)

# Identifier les types résidentiels à exclure
residential_types = [
    'Multifamily LR (1-4)',
    'Multifamily MR (5-9)', 
    'Multifamily HR (10+)',
    'Campus',
    'Residential',
    'Multifamily Housing',
    'Senior Care Community',
    'Residence Hall'
]

# Filtrer les bâtiments non résidentiels
if 'BuildingType' in building_consumption.columns:
    # Filtrer par BuildingType si disponible
    building_consumption = building_consumption[building_consumption['BuildingType'] == 'NonResidential'].copy()
    print(f"Nombre de bâtiments non résidentiels (BuildingType) : {len(building_consumption)}")
elif 'PrimaryPropertyType' in building_consumption.columns:
    # Sinon filtrer par PrimaryPropertyType dans le cas où BuildingType n'existe pas
    building_consumption = building_consumption[~building_consumption['PrimaryPropertyType'].isin(residential_types)].copy()
    print(f"Nombre de bâtiments non résidentiels (PrimaryPropertyType) : {len(building_consumption)}")

print(f"\nPourcentage conservé : {(len(building_consumption) / initial_count) * 100:.1f}%")
```

```python
# Vérifier les types de bâtiments disponibles
print("Types de bâtiments disponibles :")
if 'BuildingType' in building_consumption.columns:
    print("\nBuildingType :")
    print(building_consumption['BuildingType'].value_counts())

if 'PrimaryPropertyType' in building_consumption.columns:
    print("\n\nPrimaryPropertyType (top 20) :")
    print(building_consumption['PrimaryPropertyType'].value_counts().head(20))
```

### 2- On cherche ensuite les colonnes vides et sans intérets

```python
# On affiche toutes les colonnes disponibles avec le pourcentage de valeurs manquantes
missing_data = building_consumption.isnull().sum().sort_values(ascending=False)
missing_percentage = (missing_data / building_consumption.shape[0]) * 100
missing_percentage = missing_percentage[missing_percentage > 0]
missing_percentage.sort_values(ascending=False, inplace=True) 
missing_percentage
```

```python
# On affiche la quantité de batiment impacté par la colonne 'Outlier'
outlier_counts = building_consumption['Outlier'].value_counts()
print("Quantité de bâtiments impactés par la colonne 'Outlier' :")
for value, count in outlier_counts.items():
    print(f"- {value} : {count}")

# Nombre de bâtiments avant suppression des outliers
print(f"\nNombre de bâtiments AVANT suppression des outliers : {len(building_consumption)}")

# On supprime les batiments qui ont la colonne 'Outlier' renseignée
building_consumption = building_consumption[building_consumption['Outlier'].isna()].copy()

# Nombre de bâtiments après suppression des outliers
print(f"\nNombre de bâtiments APRÈS suppression des outliers : {len(building_consumption)}")
```

```python
# On supprime les colonnes qui ont plus de 40% de valeurs manquantes
threshold = 0.4 * len(building_consumption)
building_consumption = building_consumption.loc[:, building_consumption.isna().sum() < threshold]
print(f"\nNombre de colonnes après suppression de celles avec plus de 40% de valeurs manquantes : {building_consumption.shape[1]}")

# on affiche les colonnes restantes
building_consumption.info(verbose=True)
```

### 3. Uniformisation des unités

On uniformise les valeurs et on renomme les colonnes avec les bonnes unités

```python
# Convertion des KBtu en kWh pour uniformiser les unités énergétiques
building_consumption['SiteEnergyUse(kWh)'] = building_consumption['SiteEnergyUse(kBtu)'] * 0.29307107
building_consumption['SiteEUI(kWh/m2)'] = building_consumption['SiteEUI(kBtu/sf)'] * 3.2
building_consumption['NaturalGas(kWh)'] = building_consumption['NaturalGas(therms)'] * 29.3001
```

### 4. Explication des colonnes principales

Analysons les colonnes les plus pertinentes pour notre étude sur la consommation énergétique des bâtiments :

```python
# Identification des types de colonnes
print("Types de données par catégorie :\n")

# Colonnes d'identification
id_cols = ['OSEBuildingID', 'DataYear', 'BuildingType', 'PrimaryPropertyType', 
           'PropertyName', 'TaxParcelIdentificationNumber', 'Location']
print(f"Colonnes d'identification ({len(id_cols)}) : {id_cols}\n")

# Colonnes de superficie
area_cols = [col for col in building_consumption.columns if 'floor' in col.lower() or 'area' in col.lower()]
print(f"Colonnes du nombre d'étages ({len(area_cols)}) : {area_cols}\n")

# Colonnes de consommation énergétique
energy_cols = [col for col in building_consumption.columns if 'energy' in col.lower() or 'EUI' in col or 'kBtu' in col]
print(f"Colonnes d'énergie ({len(energy_cols)}) : {energy_cols}\n")

# Colonnes d'émissions
emission_cols = [col for col in building_consumption.columns if 'emissions' in col.lower() or 'GHG' in col]
print(f"Colonnes d'émissions ({len(emission_cols)}) : {emission_cols}")
```

### 4. Sélection des colonnes pertinentes

Pour notre analyse de la consommation énergétique, nous allons conserver les colonnes suivantes :

**Colonnes d'identification et caractérisation :**
- `OSEBuildingID` : Identifiant unique du bâtiment
- `BuildingType` : Type de bâtiment (résidentiel, commercial, etc.)
- `PrimaryPropertyType` : Sous-type plus précis du bâtiment
- `YearBuilt` : Année de construction (impact sur l'isolation et l'efficacité énergétique)

**Colonnes de superficie :**
- `PropertyGFATotal` : Surface totale du bâtiment en pieds carrés
- `PropertyGFABuilding(s)` : Surface des bâtiments
- `NumberofFloors(s)` : Nombre d'étages

**Colonnes de consommation énergétique :**
- `SiteEnergyUse(kBtu)` : Consommation énergétique totale sur site
- `SiteEUI(kBtu/sf)` : Intensité énergétique (consommation par surface)
- `Electricity(kWh)` : Consommation électrique
- `NaturalGas(therms)` : Consommation de gaz naturel

**Colonnes d'émissions :**
- `TotalGHGEmissions` : Émissions totales de gaz à effet de serre
- `GHGEmissionsIntensity` : Intensité des émissions

### 5. Nettoyage des données - Suppression des lignes et colonnes

- Suppression des lignes sans consommation énergétique (variable cible principale)
- Suppression des lignes sans surface (nécessaire pour calculer l'intensité énergétique)
- Élimination des valeurs nulles ou négatives (données aberrantes)

```python
# Analyse des valeurs manquantes dans le dataset réduit
print("Valeurs manquantes par colonne :")
missing_analysis = pd.DataFrame({
    'Colonne': building_consumption.columns,
    'Valeurs_manquantes': building_consumption.isnull().sum().values,
    'Pourcentage': (building_consumption.isnull().sum().values / len(building_consumption) * 100).round(2)
})
print(missing_analysis.to_string(index=False))

print(f"\n\nNombre de lignes initial : {len(building_consumption)}")

# Suppression des lignes avec des valeurs manquantes critiques
# On supprime les lignes sans consommation énergétique (objectif principal)
building_consumption_clean = building_consumption.dropna(subset=['SiteEnergyUse(kBtu)'])
print(f"Après suppression des lignes sans SiteEnergyUse : {len(building_consumption_clean)}")

# Suppression des lignes sans surface (impossible de calculer l'intensité)
building_consumption_clean = building_consumption_clean.dropna(subset=['PropertyGFATotal'])
print(f"Après suppression des lignes sans PropertyGFATotal : {len(building_consumption_clean)}")

# Suppression des valeurs aberrantes (énergie ou surface = 0)
building_consumption_clean = building_consumption_clean[(building_consumption_clean['SiteEnergyUse(kBtu)'] > 0) & (building_consumption_clean['PropertyGFATotal'] > 0)]
print(f"Après suppression des valeurs nulles ou négatives : {len(building_consumption_clean)}")

# Suppression des colonnes qui nous intéressent pas pour l'analyse
columns_necessary = [
    'OSEBuildingID',
    'PrimaryPropertyType',
    'YearBuilt',
    'PropertyGFATotal',
    'PropertyGFABuilding(s)',
    'NumberofFloors',
    'SiteEnergyUse(kWh)',
    'SiteEUI(kWh/m2)',
    'Electricity(kWh)',
    'NaturalGas(kWh)',
    'TotalGHGEmissions',
    'GHGEmissionsIntensity'
]

# On ne garde que les colonnes intéressantes
building_consumption_clean = building_consumption_clean[columns_necessary].copy()

print(f"\nNombre final de colonnes : {building_consumption_clean.shape[1]}")
print(f"Nombre final de lignes : {building_consumption_clean.shape[0]}")
```

```python
# On sauvegarde le jeu de données nettoyé
building_consumption_clean.to_csv("./sources/2016_Building_Energy_Benchmarking_01_cleaned.csv", index=False)

# On affiche les caractéristiques final du jeu de données nettoyé
display(building_consumption_clean.head().T)
```
