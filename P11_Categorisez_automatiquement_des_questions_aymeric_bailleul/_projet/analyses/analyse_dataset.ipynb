## 6. Definition du perimetre du projet

### 6.1 Analyse par region

```python
# Analyse des evenements par region
from datetime import timedelta

print("Repartition par region:")
print("="*60)
evenements_par_region = df['location_region'].value_counts().head(15)
for nom_region, nombre_evenements in evenements_par_region.items():
    print(f"{nom_region:30s} : {nombre_evenements:8,} evenements")
```

### 6.2 Focus sur l'Occitanie

```python
# Filtrage sur l'Occitanie
occitanie_df = df[df['location_region'].str.contains('Occitanie', case=False, na=False)].copy()

print("Statistiques Occitanie:")
print("="*60)
print(f"Total evenements Occitanie : {len(occitanie_df):,}")
print(f"Pourcentage du dataset     : {len(occitanie_df)/len(df)*100:.2f}%")
```

### 6.3 Filtrage temporel : 1 an arriere + tous futurs

```python
# Definition de la periode
today = pd.Timestamp.now(tz='Europe/Paris')
one_year_ago = today - timedelta(days=365)

print("Periode selectionnee:")
print("="*60)
print(f"Date de reference      : {today.strftime('%Y-%m-%d')}")
print(f"Debut periode (1 an)   : {one_year_ago.strftime('%Y-%m-%d')}")
print(f"Fin periode            : Tous evenements futurs")

# Filtrage temporel
occitanie_past = occitanie_df[
    (occitanie_df['firstdate_begin'] >= one_year_ago) & 
    (occitanie_df['firstdate_begin'] < today)
]

occitanie_future = occitanie_df[
    occitanie_df['firstdate_begin'] >= today
]

df_final = pd.concat([occitanie_past, occitanie_future])

print(f"\nResultats du filtrage:")
print("="*60)
print(f"Evenements 1 an arriere : {len(occitanie_past):,}")
print(f"Evenements futurs       : {len(occitanie_future):,}")
print(f"TOTAL FINAL             : {len(df_final):,} evenements")
```

### 6.4 Repartition geographique du perimetre final

## 8. Nettoyage des donnees

### 8.1 Suppression des colonnes vides

```python
# Creation d'une copie pour le nettoyage
df_clean = df_final.copy()

print(f"Dataset initial: {df_clean.shape}")
print()

# Identification et suppression des colonnes totalement vides
colonnes_a_supprimer = [nom_colonne for nom_colonne in df_clean.columns 
                        if df_clean[nom_colonne].notna().sum() == 0]

if colonnes_a_supprimer:
    print(f"Colonnes a supprimer ({len(colonnes_a_supprimer)}):")
    for nom_colonne in colonnes_a_supprimer:
        print(f"  - {nom_colonne}")
    df_clean = df_clean.drop(columns=colonnes_a_supprimer)
    print(f"\nDataset apres suppression: {df_clean.shape}")
else:
    print("Aucune colonne vide a supprimer")
```

### 8.2 Nettoyage du HTML dans longdescription_fr

```python
import re
from html import unescape

def clean_html(text):
    """Nettoie le HTML d'un texte"""
    if pd.isna(text):
        return text
    
    # Decode HTML entities
    text = unescape(str(text))
    
    # Suppression des balises HTML
    text = re.sub(r'<[^>]+>', ' ', text)
    
    # Suppression des espaces multiples
    text = re.sub(r'\s+', ' ', text)
    
    # Suppression des espaces en debut/fin
    text = text.strip()
    
    return text

# Nettoyage de longdescription_fr
print("Nettoyage du HTML dans longdescription_fr...")
if 'longdescription_fr' in df_clean.columns:
    # Echantillon avant nettoyage
    sample_before = df_clean[df_clean['longdescription_fr'].notna()].iloc[0]['longdescription_fr']
    print("\nEchantillon AVANT:")
    print(f"{sample_before[:200]}...")
    
    df_clean['longdescription_fr'] = df_clean['longdescription_fr'].apply(clean_html)
    
    # Echantillon après nettoyage
    sample_after = df_clean[df_clean['longdescription_fr'].notna()].iloc[0]['longdescription_fr']
    print("\nEchantillon APRES:")
    print(f"{sample_after[:200]}...")
    print("\n✓ Nettoyage HTML termine")
else:
    print("Colonne longdescription_fr non trouvee")
```

### 8.3 Creation d'un champ texte consolide pour le RAG

```python
# Creation d'un champ texte consolide pour la vectorisation
def create_combined_text(row):
    """Combine les champs texte pour le RAG"""
    parts = []
    
    # Titre
    if pd.notna(row['title_fr']):
        parts.append(f"Titre: {row['title_fr']}")
    
    # Description courte
    if pd.notna(row['description_fr']):
        parts.append(f"Description: {row['description_fr']}")
    
    # Description longue
    if pd.notna(row['longdescription_fr']) and len(str(row['longdescription_fr'])) > 10:
        parts.append(f"Details: {row['longdescription_fr']}")
    
    # Mots-cles
    if pd.notna(row.get('keywords_fr')):
        parts.append(f"Mots-cles: {row['keywords_fr']}")
    
    # Localisation
    if pd.notna(row['location_city']):
        parts.append(f"Lieu: {row['location_city']}")
    
    return " | ".join(parts)

print("Creation du champ texte consolide...")
df_clean['text_for_rag'] = df_clean.apply(create_combined_text, axis=1)

# Statistiques
print(f"\nLongueur moyenne du texte RAG: {df_clean['text_for_rag'].str.len().mean():.0f} caracteres")
print(f"Longueur min: {df_clean['text_for_rag'].str.len().min()}")
print(f"Longueur max: {df_clean['text_for_rag'].str.len().max()}")

# Echantillon
print("\nEchantillon de texte RAG:")
print("="*70)
print(df_clean['text_for_rag'].iloc[0])
```

### 8.4 Normalisation des dates

```python
# Verification des formats de dates
print("Verification des dates:")
print("="*70)

colonnes_dates = ['firstdate_begin', 'lastdate_end']
for nom_colonne_date in colonnes_dates:
    if nom_colonne_date in df_clean.columns:
        print(f"\n{nom_colonne_date}:")
        print(f"  Type: {df_clean[nom_colonne_date].dtype}")
        print(f"  Valeurs presentes: {df_clean[nom_colonne_date].notna().sum():,}")
        print(f"  Valeurs uniques: {df_clean[nom_colonne_date].nunique():,}")
        
        # Echantillon de dates
        echantillon_dates = df_clean[df_clean[nom_colonne_date].notna()][nom_colonne_date].head(3)
        print(f"  Echantillon:")
        for date_exemple in echantillon_dates:
            print(f"    - {date_exemple}")
```

### 8.5 Sauvegarde du dataset nettoye

```python
# Creation du repertoire processed si necessaire
processed_dir = Path("../data/processed")
processed_dir.mkdir(parents=True, exist_ok=True)

# Sauvegarde du dataset nettoye
output_path = processed_dir / "evenements_occitanie_clean.parquet"
df_clean.to_parquet(output_path, index=False)

print("Dataset nettoye sauvegarde:")
print("="*70)
print(f"Chemin: {output_path}")
print(f"Taille: {output_path.stat().st_size / (1024*1024):.2f} MB")
print(f"Lignes: {len(df_clean):,}")
print(f"Colonnes: {len(df_clean.columns)}")

# Resume des transformations
print("\nResume des transformations:")
print("="*70)
print(f"Dataset initial (filtre)    : {len(df_final):,} lignes, {len(df_final.columns)} colonnes")
print(f"Dataset nettoye             : {len(df_clean):,} lignes, {len(df_clean.columns)} colonnes")
print(f"Colonnes supprimees         : {len(df_final.columns) - len(df_clean.columns)}")
print(f"Nouvelle colonne text_for_rag ajoutee")
print(f"HTML nettoye dans longdescription_fr")
```

### 8.6 Verification finale de la qualite

Verification que le nettoyage a bien ete effectue et que les donnees sont pretes pour la vectorisation.

```python
# Verification finale de la qualite des donnees
print("VERIFICATION FINALE DE LA QUALITE")
print("="*70)

# Verification du champ RAG
print("\n1. Champ text_for_rag:")
nombre_textes_invalides = (df_clean['text_for_rag'].isna() | (df_clean['text_for_rag'].str.len() < 20)).sum()
nombre_textes_valides = len(df_clean) - nombre_textes_invalides
print(f"   Textes valides: {nombre_textes_valides:,}/{len(df_clean):,}")
print(f"   Textes trop courts/vides: {nombre_textes_invalides}")

# Verification des champs essentiels
print("\n2. Champs essentiels:")
champs_essentiels_verification = ['title_fr', 'description_fr', 'firstdate_begin', 'location_city']
for nom_champ in champs_essentiels_verification:
    nombre_valeurs_manquantes = df_clean[nom_champ].isna().sum()
    nombre_valeurs_valides = len(df_clean) - nombre_valeurs_manquantes
    pourcentage_completude = (1 - nombre_valeurs_manquantes / len(df_clean)) * 100
    print(f"   {nom_champ:20s}: {nombre_valeurs_valides:,} valides ({pourcentage_completude:.1f}%)")

# Verification du nettoyage HTML
print("\n3. Nettoyage HTML:")
if 'longdescription_fr' in df_clean.columns:
    nombre_balises_html_restantes = df_clean['longdescription_fr'].fillna('').str.contains('<[^>]+>', regex=True).sum()
    print(f"   Textes avec balises HTML restantes: {nombre_balises_html_restantes}")
else:
    print("   Colonne longdescription_fr non presente")

print("\n✓ Dataset pret pour la Phase 3 (Vectorisation)")
```
